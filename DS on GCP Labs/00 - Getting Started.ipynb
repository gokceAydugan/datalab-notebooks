## Unzip file to csv

```python
%%bash

cd data

echo ${YEAR:=2015}  # default if YEAR not set
for month in `seq -w 1 12`; do 
   unzip $YEAR$month.zip
   mv *ONTIME.csv $YEAR$month.csv
   rm $YEAR$month.zip
done
```

## Remove quotes

```python
%%bash

cd data

echo ${YEAR:=2015}  # default if YEAR not set
for month in `seq -w 1 12`; do
    echo $YEAR$month.csv
    sed 's/,$//g' $YEAR$month.csv | sed 's/"//g' > tmp
    mv tmp $YEAR$month.csv
done
```

## Upload to GCS

```python
%%bash

cd data

BUCKET=ksalama-gcs-demo
gsutil -m cp *.csv gs://$BUCKET/data/flights/raw

```
